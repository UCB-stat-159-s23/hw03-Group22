The analysis was then performed using Matlab to process scientific data. To make Figure 1, here are the following steps:
- Download 4 TB of data from NASA PO.DAAC data archive via FTP
- Go through each day of data and subset to the West Coast Region to reduce size and save each subsetted day
- Go through 2002-2012 and create a daily climatology and save all 365 days of the climatology
- Go through each day of data and calculate the anomaly and save each day's anomaly

This whole process took about 1-2 month. 
Below we will do this using MUR SST data on AWS Open Data Program in a few minutes using Python.

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import fsspec
import matplotlib.pyplot as plt

import xarray as xr
xr.set_options(display_style="html")  # display dataset nicely

import warnings
warnings.simplefilter("ignore")  # filter some warning messages

# code features from https://scitools.org.uk/cartopy/docs/v0.14/matplotlib/feature_interface.html
crs = ccrs.PlateCarree()  # set projection
```

```python
ds = xr.open_dataset("~/shared/climate-data/ds_hw.nc")
ds
```

```python
ds.analysed_sst[0, :, :].plot()

# Plot on a map
ax = plt.subplot(projection=crs)
ds.analysed_sst[0, :, :].plot(ax=ax, transform=crs)
ax.coastlines("10m", color="k")
ax.add_feature(cfeature.LAND, color="grey")
ax.add_feature(cfeature.STATES.with_scale("10m"))
```

```python
# Transform temeprature from Kkelvin to C
ds.update({"analysed_sst_C": (('time', 'lat', 'lon'), ds.analysed_sst.values - 274)})

# Calculate the climatorogy daily data from 2002 and 2012
clima_ds = ds.sel(time = slice('2002', '2012')).groupby('time.dayofyear').mean()
clima_ds
```

```python
# Calculate the standard deviation from the climatology daily data
ds.analysed_sst_C.sel(time = slice('2002', '2012')).groupby('time.month').std()[0, :, :]
```

```python
#### Check for 2014-01 data for
# Calculate anomolies by substracting 
ds_2014_value = ds.sel(time = '2014').analysed_sst_C.values - clima_ds.analysed_sst_C.values[0:-1]
ds_2014 = ds.sel(time = '2014')
ds_2014.update({"deviation_sst_C": (('time', 'lat', 'lon'), ds_2014_value)})

# Change deviation within sd < 1 to 0
a = (abs(ds_2014.sel(time = '2014-04').deviation_sst_C.mean('time')) > ds.analysed_sst_C.sel(time = slice('2002', '2012')).groupby('time.month').std()[3, :, :])*ds_2014.sel(time = '2014-04').deviation_sst_C.mean('time')

# Make plot
a.plot()
a.plot.contour(colors='black', linewidths=1)
ax = plt.subplot(projection=crs)
a.plot(ax=ax, transform=crs)
ax.coastlines("10m", color="k")
ax.add_feature(cfeature.LAND, color="grey")
ax.add_feature(cfeature.STATES.with_scale("10m"))
```

## 2. The West Coast Heat Wave
​
In 2016 [Dr. Chelle Gentemann](https://cgentemann.github.io) and collaborators published a [paper](https://agupubs.onlinelibrary.wiley.com/doi/10.1002/2016GL071039) on the heat wave in the ocean off the California coast
​
The analysis was then performed using Matlab to process scientific data. To make Figure 1, here are the following steps:
- Download 4 TB of data from NASA PO.DAAC data archive via FTP
- Go through each day of data and subset to the West Coast Region to reduce size and save each subsetted day
- Go through 2002-2012 and create a daily climatology and save all 365 days of the climatology
- Go through each day of data and calculate the anomaly and save each day's anomaly
​
This whole process took about 1-2 month. 
Below we will do this using MUR SST data on AWS Open Data Program in a few minutes using Python.

```python
# Create month
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Create subplots for the period from 2014 - 2016
fig, axs = plt.subplots(figsize=(6,25), nrows=12, ncols=3, subplot_kw=dict(projection=crs))

# Calculate the standard deviation from the climatology daily data
sd = ds.analysed_sst_C.sel(time = slice('2002', '2012')).groupby('time.month').std()

# Creating subplots for each month of year using nested for loop
for col in range(3):
    
    for row in range(12):
        
        # For year with 365 days
        if col == 2:
            values = ds.sel(time = "{}".format(col+2014)).analysed_sst_C.values - clima_ds.analysed_sst_C.values
        else:
            values = ds.sel(time = "{}".format(col+2014)).analysed_sst_C.values - clima_ds.analysed_sst_C.values[0:-1]
        temp = ds.sel(time = "{}".format(col+2014))
        temp = temp.update({"deviation_sst_C": (('time', 'lat', 'lon'), values)})
        
        # Change deviation within sd < 1 to 0
        a = (abs(temp.sel(time = "{}-{}".format(col+2014,row+1)).deviation_sst_C.mean('time')) > 
             sd[row, :, :]) * \
            temp.sel(time = "{}-{}".format(col+2014,row+1)).deviation_sst_C.mean('time')
        
        # Make subplot
        im = a.plot(ax=axs[row, col], transform=crs, add_colorbar=False, cmap = 'coolwarm', vmin = -3, vmax = 3)
        
        # Make contour plot
        a.plot.contour(ax=axs[row, col], colors='black', linewidths=1)
        axs[row, col].set_yticks(range(35,46,5))
        axs[row, col].set_ylabel('')
        
        # Add title
        if row==0:
            axs[row, col].set_title(str(col+2014)) 
            
        # Add ticks
        if row==11:
            axs[row, col].set_xticks(range(-130, -119, 10))
            axs[row, col].set_xlabel('')
        
        axs[row, col].text(-123, 45, month[row], fontsize=12, color = 'white')
        axs[row, col].coastlines("10m", color="k")
        axs[row, col].add_feature(cfeature.LAND, color="grey")
        axs[row, col].add_feature(cfeature.STATES.with_scale("10m"))
        
        

# add ticks and labels to the color bar
cfig = plt.colorbar(im, ax=axs, location='bottom', shrink=0.7, pad=0.03)
cfig.set_label('SST (°C)')
plt.show()
```
