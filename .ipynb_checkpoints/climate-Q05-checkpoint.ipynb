```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import fsspec

import xarray as xr
xr.set_options(display_style="html")  # display dataset nicely

import warnings
warnings.simplefilter("ignore")  # filter some warning messages

# code features from https://scitools.org.uk/cartopy/docs/v0.14/matplotlib/feature_interface.html
crs = ccrs.PlateCarree()  # set projection

ds = xr.open_dataset("~/shared/climate-data/ds_hw.nc")
ds
```

```python
from matplotlib import pyplot as plt
```

```python
#Segment out the west
xlat1, xlat2 = 33, 47
xlon1, xlon2 = (
    -140, 
    -118,
)
```

```python
west = ds.sel(lat = slice(xlat1, xlat2), lon = slice(xlon1, xlon2))
```

```python
#Specify time frame for 2002 to July 2016
west_2002_to_2016 = west.sel(time=slice('2002', '2016-07-31'))

#Find the max analysed sst 
west_2002_to_2016_max = west_2002_to_2016.groupby("time.dayofyear").max().analysed_sst

#Find the mean analysed sst for 2002 to 2012 (climatology)
west_2002_to_2012 = west.sel(time=slice('2002', '2012'))
west_2002_to_2012_mean = west_2002_to_2012.groupby("time.dayofyear").mean().analysed_sst

#Finding the SST anomaly
anomaly = (west_2002_to_2016_max - west_2002_to_2012_mean)
anomaly_max = anomaly.max("dayofyear")
```

```python
#Creating the plot
plt.rcParams["figure.figsize"] = (8, 6)
crs = ccrs.PlateCarree()
fig, ax = plt.subplots(subplot_kw={'projection': crs})
im = anomaly_max[:,:].plot.imshow(ax=ax, transform=crs,cmap = "jet",vmin = 0, vmax = 5)
ax.coastlines("10m", color="k")
ax.add_feature(cfeature.LAND, color="grey")
ax.add_feature(cfeature.STATES.with_scale("10m"))
gridliner = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=False)
gridliner.xlabels_bottom = True
gridliner.ylabels_left = True
gridliner.xlines = False
gridliner.ylines = False


cbar = plt.colorbar(im ,ax=ax, orientation='horizontal', location = 'bottom')
cbar.set_label("SST (C)")
plt.show()
```
